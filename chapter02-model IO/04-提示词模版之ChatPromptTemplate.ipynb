## 提示词模版之ChatPromptTempla的使用

#### 1、实例化的方式（两种方式：使用构造方法、from_messages()）
#### 2、调用提示词模板的几种方法：invoke()\format()\format_messages()\format_prompt()
* invoke()：传入的是字典，返回<class 'langchain_core.prompt_values.ChatPromptValue'>
* format()：传入变量的值，返回str
* format_messages()：传入变量的值,返回值<class 'list'>
* format_prompt()：传入变量的值,返回值<class 'langchain_core.prompt_values.ChatPromptValue'>
#### 3、更丰富的实例化参数类型
#### 4、结合LLM
#### 5、插入消息列表：MessagePlaceholder

##### 1、实例化的方式

```python
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage
# 方式1:使用构造方法
from langchain_core.prompts import ChatPromptTemplate
template=ChatPromptTemplate(
    messages=[
        ("system", "你是一个{role}专家"),
        ("human", "{question}")
    ],
    input_variables=["role", "question"]  # 显式声明
)
messages=template.invoke({'role':'物理', 'question':'请解释一下量子纠缠'})
print(messages)
print(type(messages))
```

```python
# 方式1:简易写法
from langchain_core.prompts import ChatPromptTemplate
template=ChatPromptTemplate(
    [
        ("system", "你是一个{role}专家"),
        ("human", "{question}")
    ],
)
messages=template.invoke({'role':'物理', 'question':'请解释一下量子纠缠'})
print(messages)
```

```python
# 方式2:调用方法的写法
from langchain_core.prompts import ChatPromptTemplate
template=ChatPromptTemplate.from_messages([
        ("system", "你是一个{role}专家"),
        ("human", "{question}")
    ]
    # input_variables=["role", "question"]  # 显式声明
)
messages=template.invoke({'role':'物理', 'question':'请解释一下量子纠缠'})
print(messages)
print(type(messages)) #<class 'langchain_core.prompt_values.ChatPromptValue'>
```

```python
# 方式二：使用format
from langchain_core.prompts import ChatPromptTemplate
template=ChatPromptTemplate.from_messages([
        ("system","你是一个{role}专家"),
        ("human", "{question}")
    ]
)
messages=template.format(role='物理', question='请解释一下量子纠缠')
print(messages)
print(type(messages))  # <class 'str'>
```

```python
# 方式三：使用format_messages
from langchain_core.prompts import ChatPromptTemplate
template=ChatPromptTemplate.from_messages([
        ("system","你是一个{role}专家"),
        ("human", "{question}")
    ]
)
messages=template.format_messages(role='物理', question='请解释一下量子纠缠')
print(messages)
print(type(messages))
```

```python
# 方式四：使用format_prompt()
from langchain_core.prompts import ChatPromptTemplate
template=ChatPromptTemplate.from_messages([
        ("system","你是一个{role}专家"),
        ("human", "{question}")
    ]
)
messages=template.format_prompt(role='物理', question='请解释一下量子纠缠')
print(messages)
print(type(messages))
```

```python
# 如何实现ChatPromptValue与list[messages]、字符串之间的转换
from langchain_core.prompts import ChatPromptTemplate
template=ChatPromptTemplate.from_messages([
        ("system","你是一个{role}专家"),
        ("human", "{question}")
    ]
)
messages=template.format_prompt(role='物理', question='请解释一下量子纠缠')
# 转成list
messages_list=messages.to_messages()
print(messages_list)
print(type(messages_list))

# 转成str
messages_str=messages.to_string()
print(messages_str)
print(type(messages_str))
```

## 更丰富的实例化参数类型

```python
# 举例一：元组列表类型ChatPromptTemplate.from_messages([])
from langchain_core.prompts import ChatPromptTemplate
# 定义模板
template = ChatPromptTemplate.from_messages([
    ("system", "你是一个{role}专家，回答要{style}。"),
    ("human", "请解释一下{concept}"),
    ("ai", "好的，我来解释..."),   # 可选的历史消息
    ("human", "{question}")        # 用户继续追问
])

# 使用模板生成消息列表
messages = template.format_messages(
    role="物理",
    style="通俗易懂",
    concept="量子纠缠",
    question="那它有什么实际应用？"
)

# messages 现在是可直接传给模型的消息列表
print(messages)
print(type(messages))
```

```python
# 举例二：字符串类型ChatPromptTemplate.from_messages([""])
from langchain_core.prompts import ChatPromptTemplate
# 定义模板
template = ChatPromptTemplate.from_messages([
     "你是一个{role}专家，回答要{style}。请解释一下{concept}"
])

# 使用模板生成消息列表
messages = template.format_messages(
    role="物理",
    style="通俗易懂",
    concept="量子纠缠",
)

# messages 现在是可直接传给模型的消息列表
print(messages)
print(type(messages))
```

```python
# 举例三：字典类型ChatPromptTemplate.from_messages([{}])
from langchain_core.prompts import ChatPromptTemplate
# 定义模板
template = ChatPromptTemplate.from_messages([
    {"role":"system","content":"我是一个人工智能助手，我的名字叫{name}"},
    {"role":"human","content":"我的问题是{question}"}
])

# 使用模板生成消息列表
messages = template.format_messages(
    name="小智",
    question="地球的直径是多少",
)
# messages 现在是可直接传给模型的消息列表
print(messages)
print(type(messages))
```

```python
# 举例四：Chat提示词模板类型
from langchain_core.prompts import ChatPromptTemplate
# 定义模板
nes1=ChatPromptTemplate.from_messages([
    ("system","我是一个人工智能助手，我的名字叫{name}")
])
nes2=ChatPromptTemplate.from_messages([
    ("human","我的问题是{question}")
])
template = ChatPromptTemplate.from_messages([
   nes1,nes2
])

# 使用模板生成消息列表
messages = template.format_messages(
    name="小智",
    question="地球的直径是多少",
)
# messages 现在是可直接传给模型的消息列表
print(messages)
print(type(messages))
```

```python
# 举例五：消息提示词模板类型
from langchain_core.prompts import ChatPromptTemplate,SystemMessagePromptTemplate,HumanMessagePromptTemplate
# 定义模板

str1="我是一个人工智能助手，我的名字叫{name}"
nes1=SystemMessagePromptTemplate.from_template(
    str1
)
print(nes1)
print('-------')
str1="我的问题是{question}"
nes2=HumanMessagePromptTemplate.from_template(
    str1
)

template = ChatPromptTemplate.from_messages([
   nes1,nes2
])

# 使用模板生成消息列表
messages = template.format_messages(
    name="小智",
    question="地球的直径是多少",
)
print('--------')
# messages 现在是可直接传给模型的消息列表
print(messages)
print(type(messages))
```

### 结合LLM

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
import os
import dotenv

dotenv.load_dotenv()
os.environ["OPENAI_API_KEY"]=os.getenv("OPENAI_API_KEY2")
os.environ["OPENAI_BASE_URL"]=os.getenv("OPENAI_BASE_URL2")

chat=ChatOpenAI(
    model='tongyi-xiaomi-analysis-flash',
    temperature=0.6
)
chat_prompt=ChatPromptTemplate.from_messages([
    ("system","你是一个AI助手，你的名字叫{name}"),
    ("user","我的问题是{question}")
])
chat_p_f=chat_prompt.format_messages(name='小智',question='地球的直径是多少')
llm=chat.invoke(chat_p_f)
print(llm.content)
print(type(llm))
```

### 插入消息列表：MessagePlaceholder

```python
# 举例1
from langchain_core.prompts import ChatPromptTemplate,MessagesPlaceholder
from langchain_core.messages import HumanMessage

chat_prompt=ChatPromptTemplate.from_messages([
    {"role":"system","content":"你是一个AI助手，你的名字叫{name}"},
    MessagesPlaceholder(variable_name="msgs")
])
chat_prompt.invoke({'name':'小智','question':'地球到月球的距离',
                    'msgs':[HumanMessage(content='地球的直径是多少')]})
```

```python
# 举例2
from langchain_core.prompts import ChatPromptTemplate,MessagesPlaceholder
from langchain_core.messages import AIMessage
chat_prompt=ChatPromptTemplate.from_messages([
    {"role":"system","content":"你是一个AI助手，你的名字叫{name}"},
    MessagesPlaceholder(variable_name="msgs")
])
chat_prompt.invoke({'name':'小智','question':'地球到月球的距离',
    'msgs':[HumanMessage(content='地球的直径是多少'),AIMessage(content='我需要思考一下')]
})
```

```python
# 举例3
'''
    在多轮对话中，你需要把用户和 AI 之间的历史消息传递给模型，让模型理解上下文。如果手动拼接历史，代码会很冗长。MessagesPlaceholder
    让你在模板中用一行声明，然后在调用时传入历史消息列表即可。
'''
from langchain_core.prompts import ChatPromptTemplate,MessagesPlaceholder
from langchain_core.messages import AIMessage


prompt=ChatPromptTemplate.from_messages([
    ('system','你是一个百科助手'),
    MessagesPlaceholder(variable_name='history'),
    ('human','{question}')
])
prompt_format_messages=prompt.format_messages(history=[HumanMessage(content='你的名字是什么'),AIMessage(content='工藤新一')]
                       ,question='我刚才问题是什么')

```

```python
from langchain_openai import ChatOpenAI
import os
import dotenv

dotenv.load_dotenv()
os.environ["OPENAI_API_KEY"]=os.getenv("OPENAI_API_KEY2")
os.environ["OPENAI_BASE_URL"]=os.getenv("OPENAI_BASE_URL2")

chat=ChatOpenAI(
    model='tongyi-xiaomi-analysis-flash',
    temperature=0.6
)

llm=chat.invoke(prompt_format_messages)
print(llm.content)
print(type(llm))
```
