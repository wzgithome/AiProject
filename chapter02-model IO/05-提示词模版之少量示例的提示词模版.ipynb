### 1、FewShotPromptTemplate的使用

```python
# 举例1:为提供示例的情况
from langchain_core.prompts import FewShotPromptTemplate
import os
import dotenv
from langchain_openai import ChatOpenAI

dotenv.load_dotenv()
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY2")
os.environ["OPENAI_BASE_URL"] = os.getenv("OPENAI_BASE_URL2")

chat = ChatOpenAI(
    model='tongyi-xiaomi-analysis-flash',
    temperature=0.6
)

llm = chat.invoke('2 🐔 9是多少')  # AI是不知道答案的
print(llm.content)
```

```python
# 举例2:提供示例的情况
from langchain_core.prompts import PromptTemplate, FewShotPromptTemplate

example_prompt = PromptTemplate.from_template(
    template="input:{input}\noutput:{output}",
)
examples = [
    {"input": "北京天气怎么样", "output": "北京市"},
    {"input": "上海天气怎么样", "output": "上海市"},
    {"input": "南京天气怎么样", "output": "南京市"},
]
few_shot_prompt = FewShotPromptTemplate(
    example_prompt=example_prompt,
    examples=examples,
    # prefix="请给出城市名称",
    suffix="input:{input}\noutput:",  # 声明在示例后面的提示词模板
    # input_variables=["input"]
)
few_shot_prompt.invoke({"input": "杭州天气怎么样"})
# prompt=few_shot_prompt.format_prompt(input= "杭州天气怎么样")
# print( prompt)


```

```python
# 调用模型后
from langchain_openai import ChatOpenAI
import dotenv
import os

dotenv.load_dotenv()
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY2")
os.environ["OPENAI_BASE_URL"] = os.getenv("OPENAI_BASE_URL2")

llm = ChatOpenAI(
    model='tongyi-xiaomi-analysis-flash',
    temperature=0.3
)
# llm.invoke(few_shot_prompt.format_prompt(input= "杭州市天气怎么样"))
res = llm.invoke(few_shot_prompt.format_prompt(input="杭州市天气怎么样"))
print(res.content)
```

```python
# 举例3
from langchain_core.prompts import PromptTemplate, FewShotPromptTemplate

example_prompt = PromptTemplate.from_template(
    "你是一个数学专家，算式：{input}值：{output}使用：{description}"
)
examples = [
    {"input": "1+1", "output": "2", "description": "加法"},
    {"input": "3-1", "output": "2", "description": "减法"},
]
few_shot_prompt = FewShotPromptTemplate(
    example_prompt=example_prompt,
    examples=examples,
    # prefix="",
    suffix="你是一个数学专家，算式：{input}值：{output}",  # 提示词模板
)
print(few_shot_prompt.format_prompt(input="5*2", output="10"))

```

```python
# 结合模型后
from langchain_openai import ChatOpenAI
import dotenv
import os

dotenv.load_dotenv()
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY2")
os.environ["OPENAI_BASE_URL"] = os.getenv("OPENAI_BASE_URL2")

llm = ChatOpenAI(
    model='tongyi-xiaomi-analysis-flash',
    temperature=0.3
)
res = llm.invoke(few_shot_prompt.format_prompt(input="6¥6", output="36"))
print(res.content)
```

## FewShotChatMessagePromptTemplate的使用

```python
# 举例1
from langchain_core.prompts import FewShotChatMessagePromptTemplate, ChatPromptTemplate

examples = [
    {'input': '1+1等于几', 'output': '2'},
    {'input': '中国的首都上？', 'output': '北京'}
]
msg_example = ChatPromptTemplate.from_messages([
    ('system', 'You are a helpful AI Assistant'),
    ("human", "{input}"),
    ("ai", "{output}"),
])
few_shot_prompt = FewShotChatMessagePromptTemplate(
    examples=examples,
    example_prompt=msg_example,
    input_variables=["input"]
)
print(few_shot_prompt.format())
```

```python
# 举例2:结合模型
from langchain_openai import ChatOpenAI
from langchain_core.prompts import (FewShotChatMessagePromptTemplate, ChatPromptTemplate)
import dotenv
import os

dotenv.load_dotenv()

examples = [
    {'input': '2🐔2', 'output': '4'},
    {'input': '3🐔2', 'output': '9'}
]
example_prompt = ChatPromptTemplate.from_messages([
    ('human', "{input}"),
    ('ai', "{output}")
])
few_shot_prompt = FewShotChatMessagePromptTemplate(
    examples=examples,
    example_prompt=example_prompt
)

final_prompt = ChatPromptTemplate.from_messages([
    ('system', '你是一位数学天才'),
    few_shot_prompt,
    ('human', "{input}")
])

os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY2")
os.environ["OPENAI_BASE_URL"] = os.getenv("OPENAI_BASE_URL2")

llm = ChatOpenAI(
    model='tongyi-xiaomi-analysis-flash',
    temperature=0.6
)

res = llm.invoke(final_prompt.format(input="5🐔2"))
print(res.content)
```

## ExampleSelector的使用 (示例选择器)

```python
from langchain_community.vectorstores import Chroma
from langchain_core.example_selectors import SemanticSimilarityExampleSelector
import os
import dotenv
from langchain_openai import OpenAIEmbeddings

dotenv.load_dotenv()

os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY2")
os.environ["OPENAI_BASE_URL"] = os.getenv("OPENAI_BASE_URL2")

# 使用向量模型
embeddings_model = OpenAIEmbeddings(
    model='qwen3-vl-rerank',
    check_embedding_ctx_length=False,
)
# 定义示例组
examples = [
    {' question': '天空是蓝色的？', ' answer': '是'},
    {' question': '中国是 biggest country in the world?', ' answer': '是'},
    {'question': '谁是华盛顿的外祖父', 'answer': 'George Washington'}
]
# 自定义示例选择器
examples_selector = SemanticSimilarityExampleSelector.from_examples(
    examples,
    # 这是用于生成嵌入类，用于衡量语音相似性
    embeddings_model,
    Chroma,
    # 这是要生成的示例数量
    k=1
)
question = '华盛顿的父亲是谁'
selected_examples = examples_selector.select_examples({'question': question})
print(f'与输入最相似的示例是：{selected_examples}')

```
