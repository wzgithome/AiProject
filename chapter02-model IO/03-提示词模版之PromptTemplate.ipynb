### 两种特殊结构的使用 （部分提示词模版的使用、组合提示词的使用）
#### 部分提示词模版的使用（重点）

````python
# 方式1：推荐
from langchain_core.prompts import PromptTemplate
# 定义多变量模版
prompt = PromptTemplate.from_template(
    template="请评价{product}的优缺点,包括{aspect1}",
    partial_variables={"product": "电脑"} # 提示模板所携带的部分变量词典。部分变量填充模板，这样你就不需要每次调用提示时都传入它们。
)
prompt_1=prompt.format(aspect1="拍照质量")
print(prompt_1)
print(type(prompt_1))

````

````python
# 方式2
from langchain_core.prompts import PromptTemplate
prompt = PromptTemplate(
    template="请评价{product}的优缺点,包括{aspect1}",
    # 提示模板所携带的变量
    input_variables=["product","aspect1"]
).partial(product="电脑")
prompt_2=prompt.format(aspect1="拍照质量")
print(prompt_2)
````

````python
# 方式3
from langchain_core.prompts import PromptTemplate
prompt = PromptTemplate.from_template(
    template="请评价{product}的优缺点,包括{aspect1}"
).partial(product="电脑")
prompt_3=prompt.format(aspect1="打游戏卡")
print(prompt_3)
````

#### 给变量赋值的两种方式：format()/invoke()
```format():参数部分：给变量赋值；返回值：str类型```
```invoke():参数部分：给变量赋值；返回值：PromptValue类型```

````python
# 举例使用format()
from langchain_core.prompts import PromptTemplate
# 定义多变量模版
prompt = PromptTemplate.from_template(
    template="请评价{product}的优缺点,包括{aspect1}"
)
prompt_1=prompt.format(product="电脑",aspect1="拍照质量")
print(prompt_1)
print(type(prompt_1)) #<class 'str'>
````

````python
# 举例使用invoke()
from langchain_core.prompts import PromptTemplate
# 定义多变量模版
prompt = PromptTemplate.from_template(
    template="请评价{product}的优缺点,包括{aspect1}"
)
# 调用invoke()
prompt_1=prompt.invoke({"product":"电脑","aspect1":"拍照质量"})
print(prompt_1)
print(type(prompt_1)) # <class 'langchain_core.prompts.prompt_value.PromptValue'>
````

#### 结合大模型的使用

````python
from langchain_core.prompts import PromptTemplate
from langchain_openai import ChatOpenAI
import os
import dotenv

dotenv.load_dotenv()
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY2")
os.environ["OPENAI_BASE_URL"] = os.getenv("OPENAI_BASE_URL2")
llm = ChatOpenAI(
    model='tongyi-xiaomi-analysis-pro',
    temperature=0.6
)
prompt = PromptTemplate.from_template(
    template="请评价{product}的优缺点,包括{aspect1}"
)
# prompt_1=prompt.invoke({"product":"笔记本电脑","aspect1":"拍照质量"})
prompt_1=prompt.format(product="笔记本电脑",aspect1="拍照质量")

res=llm.invoke(prompt_1)
print( res.content)
print(type(res))
````
