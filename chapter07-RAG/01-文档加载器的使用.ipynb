## 文档加载器

* txt文档：TextLoader
* pdf文档：PyPDFLoader
* csv文档：CSVLoader
* json文档：JSONLoader
* html文档：UnstructuredHTMLLoader
* md文档：UnstructuredMarkdownLoader
* 文件目录：DirectoryLoader

<font color='pink'>创建好XxxxLoader的实例以后，都需要调用load()，在内存中返回
一个list[Document]</font>

```python
# 加载Text文件utf-8格式的文件
from langchain_community.document_loaders import TextLoader

# 指明txt文档的路径
file_path='./asset/load/01-langchain-utf-8.txt'

# 创建一个TextLoader的案例
text=TextLoader(
    file_path=file_path,
    # 需要指导文件的编码格式
    encoding='utf-8'
)

# 调用load()，返回一个list[Document]
docs=text.load()
print(docs)
# 显示Document对象的元数据
print(docs[0].metadata)
# 显示文档中的内容信息
print(docs[0].page_content)
```

```python
# 举例1:加载JSON文件
from langchain_community.document_loaders import JSONLoader

json=JSONLoader(
    file_path='./asset/load/04-load.json',
    jq_schema='.',  #表示加载所有的字段
    text_content=False, #将加载的json对象转换为json字符串
)
docs=json.load()
print(docs)
```

```python
# 举例2:加载JSON文件里面的某个字段
from langchain_community.document_loaders import JSONLoader

json=JSONLoader(
    file_path='./asset/load/04-load.json',
    jq_schema='.messages[].content',  #表示加载messages[]的所有的content字段
    text_content=False, #将加载的json对象转换为json字符串
)
docs=json.load()
for doc in docs:
    print(doc.page_content)
```

```python
# 举例3:加载JSON文件里面的某个字段
from langchain_community.document_loaders import JSONLoader

json=JSONLoader(
    file_path='./asset/load/04-response.json',
    jq_schema='.data.items[].content',  #表示加载data.items[].content
    text_content=False, #将加载的json对象转换为json字符串
)
docs=json.load()
# print(docs)

for doc in docs:
    print(doc.page_content)
```

```python
# 举例3:第二种方式
from langchain_community.document_loaders import JSONLoader

json=JSONLoader(
    file_path='./asset/load/04-response.json',
    jq_schema='.data.items[]',  #表示加载data.items[].content
    content_key='.content',
    is_content_key_jq_parsable=True # 用jq解析content_key
)
docs=json.load()


for doc in docs:
    print(doc.page_content)
```

```python
# 举例4:获取2个字段
from langchain_community.document_loaders import JSONLoader

json=JSONLoader(
    file_path='./asset/load/04-response.json',
    jq_schema='.data.items[]',  #表示加载data.items[].content
    content_key='.content+"\\n\\n"+.title',
    is_content_key_jq_parsable=True # 用jq解析content_key
)
docs=json.load()


for doc in docs:
    print(doc.page_content)
```
