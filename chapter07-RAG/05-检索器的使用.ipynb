```python
import dotenv
import os
from langchain_text_splitters import CharacterTextSplitter
from langchain_community.embeddings import DashScopeEmbeddings
from langchain_community.document_loaders import TextLoader
# 之前使用的是Chrome
from langchain_community.vectorstores import FAISS,Chroma

dotenv.load_dotenv(dotenv_path='../chapter02-model IO/.env')

# 自定义文档加载器
loader=TextLoader(file_path='./asset/load/09-ai1.txt',encoding='utf-8')

# 加载文档
documents=loader.load()

# 自定义文本切割器
text_splitter=CharacterTextSplitter(chunk_size=1000,chunk_overlap=0)

# 切割文档
docs=text_splitter.split_documents(documents)

# 自定义模型
os.environ["DASHSCOPE_API_KEY"] = os.getenv("OPENAI_API_KEY2")
os.environ["DASHSCOPE_BASE_URL"]=os.getenv("OPENAI_BASE_URL2")
embeddings=DashScopeEmbeddings(model="text-embedding-v3")
```

```python
# 获取向量数据库
db=Chroma.from_documents(documents=docs,embedding=embeddings)

# 基于向量数据库获取检索器
retriever=db.as_retriever()

# 进行数据的检索
docs=retriever.invoke(input="深度学习是什么？")

print(len(docs))

for doc in docs:
    print(f'-----{doc}')
```

```python
# 1.导入相关依赖
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import DashScopeEmbeddings
from langchain_core.documents import Document

# 2.定义文档
document_1=Document(
    page_content="经济复苏：美国经济正在从疫情中疫情中强劲复苏，失业率降至历史低点。！"
)
document_2=Document(
    page_content="基础设施：政府将投资1万亿美元用于修复道路、桥梁和宽度网络。"
)
document_3 = Document(
    page_content="就业增长：美国5月非农就业人数增加27.2万人，超出市场预期。"
)

document_4 = Document(
    page_content="科技创新：人工智能初创企业融资额再创新高，生成式AI成为热点。"
)

document_5 = Document(
    page_content="绿色能源：太阳能和风能发电占比首次超过煤炭，清洁转型加速。"
)

document_6 = Document(
    page_content="房地产市场：30年期房贷利率降至6.5%，购房需求回暖。"
)

document_7 = Document(
    page_content="消费趋势：在线零售额同比增长12%，实体店客流量逐步恢复。"
)

document_8 = Document(
    page_content="全球贸易：供应链压力指数降至历史低位，港口吞吐量恢复正常。"
)

document_9 = Document(
    page_content="货币政策：美联储维持基准利率不变，暗示年内降息一次。"
)

document_10 = Document(
    page_content="医疗健康：新冠疫苗升级版获批，针对新变异株保护率达85%。"
)

document_11 = Document(
    page_content="教育投入：联邦政府新增200亿美元拨款，用于缩小数字鸿沟。"
)

document_12 = Document(
    page_content="交通运输：电动公交车订单量翻倍，城市公交系统加速电动化。"
)
# 3.创建向量存储
embeddings=DashScopeEmbeddings(
    model="text-embedding-v3"
)
# 将文档向量化，添加到向量数据库索引中，得到向量数据库对象
db=Chroma.from_documents(documents,embeddings)
```

```python
# 默认检索器使用相似性搜索

retriever=db.as_retriever(search_kwargs={"k":3})
docs=retriever.invoke("经济政策")

for i,doc in enumerate(docs):
    print(f"\n结果{i+1}:\n{doc.page_content}\n")

```

```python
# 分数阀值查询:只有相似度超过这个值才会召回
retriever=db.as_retriever(
    search_type="similarity_score_threshold",
    search_kwargs={"score_threshold":0.1}
)
docs=retriever.invoke("经济政策")
for doc in docs:
    print(f"内容:{doc.page_content}")
```
